### ヒント
[Q3-2](q3-2_display_text.ipynb)同様、 _set_display_text()_ を使っていきます。  

今回は _refresh=_ 引数の穴埋めをします。  
refreshを True にするとset_display_textをする前に画面全体を現在の背景色で更新します。defaultは True です。  
Falseにすると前回の表示を残したまま文字を描画することができます。  
これまで表示していた文字や画像を消さずに、続けて表示したい場合は、refresh=を False にします。

今回は最初の"あか"の時は画面を消して新しく表示し、残りの"きいろ", "あお"の時は消さずに続けて表示していくようにしましょう。

下記のセルを実行すると使い方の例が見られます。

````python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

# akari_clientをインポートする。
from akari_client import AkariClient
# 色を指定する際に使うColors,Colorのインポート
from akari_client.color import Colors, Color
# 位置を指定する際に使うPositionsのインポート
from akari_client.position import Positions
# sleep関数を使うためにtimeのライブラリをインポートする。
import time

# AkariClient、m5stackのインスタンスを取得する
akari = AkariClient()
m5 = akari.m5stack

m5.set_display_color(Colors.WHITE)

text = "新規"
pos_x = Positions.CENTER
pos_y = 30
size = 4
text_color = Colors.BLACK
back_color = Colors.WHITE
refresh = True
m5.set_display_text(
    text=text,
    pos_x=pos_x,
    pos_y=pos_y,
    size=size,
    text_color=text_color,
    back_color=back_color,
    refresh=refresh
    )
# 0.5秒停止
time.sleep(0.5)

text = "上書きしない"
pos_x = Positions.CENTER
pos_y = 150
size = 4
text_color = Colors.RED
back_color = Colors.WHITE
refresh = False
m5.set_display_text(
    text=text,
    pos_x=pos_x,
    pos_y=pos_y,
    size=size,
    text_color=text_color,
    back_color=back_color,
    refresh=refresh
    )
````

text_color,back_colorを指定することで文字色と文字の背景色を変更できます。  

色はset_display_colorと同様RGB値を指定する方法と、Colorsクラスから色名を指定する方法の両方を使うことができます。  
それぞれテキストの表示色に異なる色が指定されています。  
[akari_manual](https://akari-docs.vbcpp.net/source/spec/soft_base.html#akari_client.color.Colors)から指定の色を選んで穴埋めしてください。  

完成したら実行してみましょう。  
M5のディスプレイに「あか」「きいろ」「あお」とそれぞれの色で表示されれば成功です。
